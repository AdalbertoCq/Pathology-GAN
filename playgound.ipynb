```python
from preparation.utils import load_data
import os
import preparation.utils as utils

patch_h = 128
patch_w = 128
n_channels = 3
project_path=os.getcwd()

relative_dataset_path = os.path.join('data', 'nki_he')
dataset_path = os.path.join(project_path, relative_dataset_path)
sets_file_path = os.path.join(dataset_path, 'sets_h%s_w%s' % (patch_h, patch_w))
augmented_sets_file_path = os.path.join(dataset_path, 'augmentations_h%s_w%s' % (patch_h, patch_w))
pathes_path = os.path.join(relative_dataset_path, 'patches_h%s_w%s' % (patch_h, patch_w))
hdf5_train = os.path.join(pathes_path, 'hdf5_nki_he_train.h5')
hdf5_test = os.path.join(pathes_path, 'hdf5_nki_he_test.h5')
hdf5 = [hdf5_train, hdf5_test]

sets = load_data(sets_file_path)
augmentations = load_data(augmented_sets_file_path)

for index, s in enumerate(['train', 'test']):
    t_path = os.path.join(pathes_path, s)
    os.makedirs(t_path)
    utils.get_and_save_patch(augmentations[index], sets[index], hdf5[index], dataset_path, t_path, patch_h, patch_w, n_channels, s)
```
